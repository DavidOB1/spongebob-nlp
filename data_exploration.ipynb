# Data Exploration

Examining the Spread of Ratings

```python
import pandas as pd

sb_data = pd.read_csv("spongebob_data.csv", encoding = "utf-8")
ratings = sb_data["rating"]

colors = {
    "boxes": "Yellow",
    "whiskers": "Brown",
    "medians": "Red",
    "caps": "Blue",
}

print(ratings.describe())
ratings.plot.box(vert=False, color = colors)
```

Examining the Ratings Across the Seasons

```python
import matplotlib.pylab as plt

season_ratings = []

for i in range(1, 13):
    season_df = sb_data.loc[sb_data["season"] == i]
    season_ratings.append(season_df["rating"].mean())

plt.plot(list(range(1, 13)), season_ratings)
plt.xlabel("Season")
plt.ylabel("Average Rating")
plt.show()
```
